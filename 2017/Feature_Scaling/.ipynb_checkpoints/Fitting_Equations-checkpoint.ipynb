```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler

%matplotlib inline
```

### Creating the data set

```python
X = np.linspace(0,2,1001)
y_real = np.sin(2*np.pi*X)
y = y_real + (np.random.randn(len(X)))/5
```

```python
plt.plot(X,y_real,color='red',linewidth=5)
plt.scatter(X,y,s=2,color='blue')
```

```python
# from sklearn import linear_model
# reg = linear_model.LinearRegression()
# X=X.reshape(-1,1)
# y=y.reshape(-1,1)
# reg.fit(X,y)
# y_pred = reg.predict(X)
```

```python
def FittingLinear(X,y,deg,norm = True):
#     interc = np.ones(len(X))
#     interc=interc.reshape(-1,1)

#     X_inter = X#np.hstack((interc,X))


    X = X.reshape(-1,1)
    p = PolynomialFeatures(deg,include_bias=True) # adds the intercept column
    X_poly = p.fit_transform(X)
#     print(X_poly[0:5,:])

    if norm:
        scaler = StandardScaler()
        X_poly = scaler.fit_transform(X_poly)

    reg = linear_model.LinearRegression()
    reg.fit(X_poly,y)
    y_pred = reg.predict(X_poly)
    
    plt.plot(X,y_pred,color ='red')
    plt.scatter(X,y,color='lightblue')
    plt.plot(X,y_real,color='purple')
    
    
#     print(X[0:5])
#     print(X_poly[0:5])
```

```python
deg = 7
FittingLinear(1e6*X,y,deg,False)
```

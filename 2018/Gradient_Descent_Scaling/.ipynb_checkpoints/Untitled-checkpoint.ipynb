```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# import random

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
min_range = -10
max_range = 10
num_samples = 100
variance = 5
beta = 2
x = np.random.uniform(min_range,max_range,num_samples)
y = beta*x+np.random.normal(0,variance,num_samples)
```

```python
def draw():
    plt.figure(figsize=[10,5])
    plt.subplot(1,2,1)
    plt.scatter(x,y)
    plt.plot(x,beta*x)
    plt.xlabel('X-Value')
    plt.ylabel('Y-Value')
    plt.legend(['Scatter','Line'])
    plt.title('Scatter')
    
    plt.subplot(1,2,2)
    plt.hist(y)
    plt.xlabel('X-Value')
    plt.ylabel('Y-Value')
    plt.legend(['Scatter','Line'])
    plt.title('Histogram')
```

```python
plt.style.use('default')
draw()
```

```python
plt.style.use('default')
plt.xkcd()
draw()
```

```python
plt.style.use('default')
plt.style.use('fivethirtyeight')
draw()
```

```python
for i in plt.style.available:
    plt.style.use('default')
    plt.style.use(i)
    print(i)
    draw()
```

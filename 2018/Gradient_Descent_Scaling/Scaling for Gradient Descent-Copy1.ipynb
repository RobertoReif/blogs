```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
import time
# from IPython import display
from matplotlib import animation, rc
from IPython.display import HTML, Image

pd.set_option('display.float_format', '{:.2g}'.format)

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
num_samples = 1000
x1 = np.random.uniform(0,1e3,num_samples)
x2 = np.random.uniform(0,1e2,num_samples)
noise = np.random.normal(0,.2,num_samples)

beta1 = 2e0
beta2 = 2e0

y = beta1*x1 + beta2*x2 + noise

X = np.array([x1,x2]).T
```

```python
# fig, ax = plt.subplots()
# First set up the figure, the axis, and the plot element we want to animate
# fig, ax = plt.subplots()

fig = plt.figure(figsize=[10,10])
ax = fig.add_subplot(111, projection='3d')

ax.set_xlim(( 0, max(x1)))
ax.set_ylim((0, max(x2)))
ax.set_zlim((0, max(y)))

line, = ax.plot([], [], [], lw=2)
```

```python
def init():
#     font = {'family' : 'sans-serif',
#             'weight' : 'normal',
#             'size'   : 15}
#     plt.rc('font', **font)

#     fig = plt.figure(figsize=[10,10])
#     ax = fig.add_subplot(111, projection='3d')
    line.set_data([], [])
    return (line,)
```

```python
rc('animation', html='html5') 
```

```python
# animation function. This is called sequentially
def animate(i):
#     print(i)
#     line.scatter(X[:,0], X[:,1], y,s=5,c='b');
    line.set_data(X[:,0], X[:,1])
    line.set_3d_properties(y)
#     line.set_3d_properties?
    line.axes.view_init(30,i+250)

    # color of the pane background
#     line.w_xaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
#     line.w_yaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
#     line.w_zaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))

#     x1_t = np.linspace(min(x1),max(x1),100)
#     x2_t = np.linspace(min(x2),max(x2),100)

    # add labels
#     line.set_xlabel('$X_1$',fontdict={'size':25,'weight':'bold'})
#     line.set_ylabel('$X_2$',fontdict={'size':25,'weight':'bold'})
#     line.set_zlabel('$Y$',fontdict={'size':25,'weight':'bold'});
#     line.view_init(30, i+250)
#     fig.canvas.draw()
#     display.clear_output(wait=True)
#     display.display(plt.gcf())
    
#     line.set_data(x, y)
#     time.sleep(.1)
    return (line,)
```

```python

# for i in range(0,360,10):
#     print(i)
# call the animator. blit=True means only re-draw the parts that 
# have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=20, blit=True)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors

%matplotlib inline
```

```python
df = pd.read_csv('features.csv')
df.head()
```

```python
df.set_index('shoe_number',inplace=True)
df.head()
```

Define a KNN of 4

```python
nbrs = NearestNeighbors(n_neighbors=4,metric='euclidean').fit(df)
```

Finds the distance and index of the newrest neighbors.  Note that the first neighbor is itself.

```python
distances, indices = nbrs.kneighbors(df)
```

```python
knn_data = pd.DataFrame(indices)+1 # the +1 adds to the index and converts into shoe_number
knn_data.to_csv('knn.csv')
```

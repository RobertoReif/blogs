```python
from PIL import Image
from PIL import ImageFont
from PIL import ImageDraw
import matplotlib.pyplot as plt

%matplotlib inline
```

### Load the image and display

```python
img = Image.open('Images/Lioness.jpeg')
plt.imshow(img)
plt.axis('off');
```

### Loads the font and adds it on the image

```python
draw = ImageDraw.Draw(img)
font_loc = '/Users/robertomac/anaconda/pkgs/matplotlib-2.0.2-np113py36_0/lib/python3.6/site-packages/matplotlib/mpl-data/fonts/ttf/cmb10.ttf'
font = ImageFont.truetype(font_loc,256) # fontfile, size
draw.text((0,0),'Lioness',(255,255,255),font = font) # (x,y),"text",(R,G,B)
```

```python
plt.imshow(img)
plt.axis('off');
```

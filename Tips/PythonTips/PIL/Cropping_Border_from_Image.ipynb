```python
from PIL import Image
import PIL.ImageOps    
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

### Loads and displays the original image

```python
img = Image.open('Images/musical-note.png')
```

```python
plt.imshow(img,cmap='gray')
plt.title('Original Image');
```

### Determines the 4 box corners where there is no white 

```python
left,upper,right,bottom = img.getbbox()
print('Left:',left)
print('Right:',right)
print('Upper:',upper)
print('Bottom:',bottom)
```

### Crops and displays the cropped image

```python
img_crop = img.crop((left,upper,right,bottom))
plt.imshow(img_crop,cmap='gray')
plt.title('Original Image');
```

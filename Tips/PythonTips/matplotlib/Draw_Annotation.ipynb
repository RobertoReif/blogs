```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

Create some data

```python
np.random.seed = 7
X = pd.DataFrame(np.random.random(100)*100)
X[1] = X[0]+np.random.randn(100)*10
```

Draw an annotation

```python
plt.figure(figsize=[5,5])
plt.scatter(X[0],X[1],alpha = 0.5, c='purple');

plt.annotate('Main Point', xy=(45, 55), xytext=(20, 80),
            arrowprops=dict(facecolor='black'))
```

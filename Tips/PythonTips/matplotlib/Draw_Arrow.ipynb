```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

Create some data

```python
np.random.seed = 7
X = pd.DataFrame(np.random.random(100)*100)
X[1] = X[0]+np.random.randn(100)*10
```

Adding an arrow

```python
plt.figure(figsize=[5,5])
plt.scatter(X[0],X[1],alpha = 0.5, c='purple');
plt.arrow(35,65,10,-10,width=0.5,head_width=3,color='black')
```
